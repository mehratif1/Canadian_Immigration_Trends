```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

```python
# Read the data into a Pandas DataFrame
canada_df = pd.read_csv('1710004001_databaseLoadingData.csv')
canada_df.head()
```

```python
# Extract the year from the "REF_DATE" column
canada_df["REF_DATE"] = canada_df["REF_DATE"].str[:4]
canada_df.head()
```

```python
#Renaming the columns 'REF_DATE' and 'VALUE'
canada_df = canada_df.rename(columns={'REF_DATE':'YEAR' , 'VALUE' : 'IMMIGRANTS COUNT'})
canada_df.head()
```

```python
#  Droping the specified columns from canada_df
columns_to_drop = ['DGUID', 'Components of population growth', 'UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE',
                    'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS']

canada_df.drop(columns=columns_to_drop, inplace=True)
```

```python
canada_df.head()
```

```python
canada_df.info()
```

```python
#Dropping the null values and groupby Year to find the sum of immigrants
canada_df_cleaned = canada_df.dropna(subset=["IMMIGRANTS COUNT"])
canada_result_df = canada_df_cleaned.groupby("YEAR").agg({
    "GEO": "first", 
    "IMMIGRANTS COUNT": "sum"
}).reset_index()
```

```python
canada_result_df["IMMIGRANTS COUNT"] = canada_result_df["IMMIGRANTS COUNT"].astype(int)
```

```python
canada_result_df
```
